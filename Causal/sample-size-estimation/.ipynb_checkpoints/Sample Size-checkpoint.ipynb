```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def std_error(sigma, n): 
    return sigma / np.sqrt(n)


control_mu = 4.5
control_std = 50
control_n = 4137

control_ci = (control_mu + 1.96 * std_error(control_std, control_n),
              control_mu - 1.96 * std_error(control_std, control_n))

control_ci
```

```python
test_mu = 5.0 # ??
test_std = 50
# test_n = ???
```

```python
sizes = np.arange(1000, 1000000, 100)
test_mus = np.ones(sizes.shape) * test_mu
test_stds = np.ones(sizes.shape) * test_std

control_mus = np.ones(sizes.shape) * control_mu
control_stds = np.ones(sizes.shape) * control_std


plt.figure(figsize=(20,8))

# test
plt.plot(sizes, test_mus + 1.96*std_error(test_std, sizes), c="b", label="test") 
plt.plot(sizes, test_mus - 1.96*std_error(test_std, sizes), c="b") 

# control
plt.plot(sizes, control_mus + 1.96*std_error(control_stds, sizes), c="r", label="control")
plt.plot(sizes, control_mus - 1.96*std_error(control_stds, sizes), c="r") 

plt.legend()
plt.show()
```

```python
true_factor = np.ones(sizes.shape) * (test_mu - control_mu)

plt.figure(figsize=(20,8))

plt.plot(sizes,
         (test_mus - std_error(test_std, sizes)) - (control_mus + std_error(control_stds, sizes)),
         c="b", label="Detected Value") 


plt.plot(sizes, true_factor, label = "Hypothesised Value")

plt.legend()
plt.ylim((0, 0.6))
plt.show()
```

## Confidence Interval Formula

Difference between to normal distributions
$$(U - V) \sim N(\mu_U - \mu_V, \sigma_U^2 + \sigma_V^2)$$

## Z-Test

$$\mu_{diff} = \mu_{test} - \mu_{control}$$

$$SE_{diff} = S_p  \sqrt{\frac{1}{n_{test}} + \frac{1}{n_{control}}}$$
$$S_p = \sqrt{\frac{(n_{test} - 1)s^2_{test} + (n_{control} - 1)s^2_{control}}{n_{test} + n_{control} - 2}}$$

```python
def std_error_diff(n_test, var_test, n_control, var_control):
    num = (n_test-1)*var_test + (n_control-1)*var_control
    denom = n_test + n_control - 2
    sp = np.sqrt(num/denom)
    return sp * np.sqrt(1/n_test + 1/n_control)
```

```python
diff_mu = (test_mus - control_mus)
diff_std = std_error_diff(sizes, np.square(test_std), sizes, np.square(control_std))
```

```python
plt.figure(figsize=(20,8))

plt.plot(sizes, diff_mu, c="r", label="mean")
plt.plot(sizes, diff_mu + 1.96*diff_std, c="b", label="CI 95%")
plt.plot(sizes, diff_mu - 1.96*diff_std, c="b") 

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,8))

plt.plot(sizes, diff_mu, c="r", label="mean")
plt.plot(sizes, diff_mu - 1.96*diff_std, c="b", label="CI Lower") 

plt.ylim((0, 0.6))
plt.legend()
plt.show()
```
